```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("top10k_logos.csv")

df = df.dropna(subset=["category_groups_list"])
df_expanded = df.assign(category_groups_list=df["category_groups_list"].str.split(",")).explode("category_groups_list")
df_expanded["category_groups_list"] = df_expanded["category_groups_list"].str.strip()

counts = df_expanded["category_groups_list"].value_counts()

plt.figure(figsize=(12,6))
counts.plot(kind="bar", color="lightcoral", edgecolor="black")

plt.title("Individual Categories by Count", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
country_col = "country_code"

df_country = df.dropna(subset=[country_col])

country_counts = df_country[country_col].value_counts()

plt.figure(figsize=(10,6))
country_counts.head(20).plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Top 20 Countries by Number of Companies", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
